```python

!pip install jupyter-dash
from jupyter_dash import JupyterDash
import dash
import dash_core_components as dcc
import dash_html_components as html
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.mosaicplot import mosaic
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("womens_harassment.csv")
df.head(22)
```

```python
# Initialize the app
app = dash.Dash(__name__)

# Define the layout of the app
app.layout = html.Div([
   
    html.H1("Data Visualization Dashboard"),
   
    # Histogram
    dcc.Graph(
        id='histogram',
        figure={
            'data': [{
                'x': df['women'],
                'y': df['children'],
                'type': 'histogram'
            }],
            'layout': {
                'title': 'Histogram'
            }
        }
    ),
   
     
    # Box plot
    dcc.Graph(
        id='boxplot',
        figure={
            'data': [{
                'x': df['children'],
                'y': df['women'],
                'type': 'box'
            }],
            'layout': {
                'title': 'Box plot'
            }
        }
    ),
   
   
    # Scatter plot
    dcc.Graph(
        id='scatterplot',
        figure={
            'data': [{
                'x': df['women'],
                'y': df['children'],
                'mode': 'markers',
                'marker': {
                    'size': 10,
                    'color': 'red',
                    'opacity': 0.8
                }
            }],
            'layout': {
                'title': 'Scatter plot'
            }
        }
    ),
   
    # Bar chart
    dcc.Graph(
        id='barchart',
        figure={
            'data': [{
                'x': df['children'],
                'y': df['women'],
                'type': 'bar'
            }],
            'layout': {
                'title': 'Bar chart'
            }
        }
    ),


  # Mosaic plot
    dcc.Graph(id='mosaicplot',
              figure={
        'data': [{
            'x': df['children'],
            'y': df['women'],
            'type': 'heatmap',
            'colorscale': 'Plasama'
        }],
           'layout': {
                'title': 'Mosaic plot'
            }        
    }),
 ])



```

```python
# Run the app
if __name__ == '__main__':
    app.run_server(port=8052)
```
